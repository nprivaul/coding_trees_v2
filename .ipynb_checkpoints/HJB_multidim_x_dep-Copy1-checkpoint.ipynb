```python
import numpy as np

def simulate_mean_reverting_interacting(
    a, gamma2, sigma,
    x0,        # array of shape (M,) initial values x^i
    T, N,      # final time, number of time steps
    M=None,    # number of particles (optional if derived from x0)
    seed=0
):
    np.random.seed(seed)

    if M is None:
        M = len(x0)

    dt = T / N
    sqrt_dt = np.sqrt(dt)

    # solution array: X[k, i] = particle i at time step k
    X = np.zeros((N+1, M))
    X[0] = x0

    drift_coef = a + gamma2

    for k in range(N):
        Xbar = X[k].mean()
        dW = np.random.randn(M) * sqrt_dt
        drift = drift_coef * (Xbar - X[k])
        X[k+1] = X[k] + drift * dt + sigma * dW

    return X
```

```python
def feynman_kac_estimator_vectorized2(
    zzeta, a, gamma1, gamma2, gamma3, sigma,
    x0, T, N, n_simulations=10000, seed=0):
    """
    Vectorized version for better performance.
    """
    np.random.seed(seed)
    M = grid_d_dim.shape[0]
    dt = T / N
    prefactor1 = gamma3 / (2 * sigma**2)
    prefactor2 = (gamma1 - gamma2**2) / (2 * sigma**2)
    
    # Pre-allocate arrays
    all_X = np.zeros((n_simulations, N+1, M))
    all_X[:, 0] = x0
    
    drift_coef = a + gamma2
    
    # Generate all Brownian increments at once
    dW = np.random.randn(n_simulations, N, M) * np.sqrt(dt)
    
    time_integrals = np.zeros(n_simulations)
    
    # Time stepping
    for k in range(N):
        # Compute mean for each simulation
        Xbar_k = all_X[:, k].mean(axis=1, keepdims=True)
        
        # Drift term
        drift = drift_coef * (Xbar_k - all_X[:, k])
        
        # Update
        all_X[:, k+1] = all_X[:, k] + drift * dt + sigma * dW[:, k]
        
        # Accumulate time integral
        diff_k = Xbar_k - all_X[:, k]
        time_integrals += np.sum(diff_k**2, axis=1) * dt
    
    # Terminal time contribution
    Xbar_T = all_X[:, -1].mean(axis=1, keepdims=True)
    diff_T = Xbar_T - all_X[:, -1]
    terminal_sums = np.sum(diff_T**2, axis=1) + zzeta * np.sum(diff_T**4, axis=1)
    
    # Compute exponential terms
    total_arguments = (- gamma3 / (2 * sigma**2) * terminal_sums 
                       - prefactor2 * time_integrals)
    exp_terms = np.exp(total_arguments)
    
    phi_estimate = np.mean(exp_terms)
    std_error = np.std(exp_terms) / np.sqrt(n_simulations)
    
    return phi_estimate, std_error, exp_terms
```

```python
fixed_dim, dim = 0, 5
T = torch.tensor(0.25 , dtype=torch.float32)  # Time horizon

zzeta=0.1

gamma1=0.1; gamma2=0.1; gamma3=0.1;

mu0 = -gamma2 + 0.1; mu = mu0 + 0.0 # mu0+0.1*dim

lbd = torch.tensor(1.0, dtype=torch.float32)  # Lambda parameter

t_lo, x_lo, x_hi = 0., -2., 2.
x_mid = (x_lo + x_hi) / 2
# x0 = x_mid * np.eye(1, dim) 
# x0 = 1.0 * np.eye(1, dim) 
x0 = np.eye(1, dim, fixed_dim)[0]

u1=0.
u2=1.
nu=2
lambd = (u1 + u2) / 2

def nth_derivatives(order, y, x):
    """
    Compute nth derivative of y with respect to x.
    """
    for _ in range(order):
        try:
            grads = torch.autograd.grad(y.sum(), x, create_graph=True)[0]
        except RuntimeError:
            # Handle zero gradient case gracefully
            return torch.zeros_like(y)
        y = grads
    return y

def u_cv(x, T):
    return u2+(u1-u2)/(1+torch.exp((u1-u2)*(mu0*x.sum(dim=0)-mu0**2*(u1+u2)*T*dim/2)/nu))                       

def u_cv_der(y, T):
    """
    First derivative of u_cv with respect to y.
    """
    return nth_derivatives(1, u_cv(y, T), y)

def f_cv0(y, T, mu):
    u_prime = u_cv_der(y, T)
    return (
        - mu * u_cv(y, T) * u_prime
    )
    
def phi(x,mu):
    """
    Example phi function.
    """
    return u2+(u1-u2)/(1+torch.exp((u1-u2)*(mu0*x.sum(dim=0))/nu))                       

def cov(x0, T, mu1, mu2): 
    """
    Compute covariance.
    """
    # Define the integrand for numerical integration
    def integrand(s):
        s_tensor = torch.tensor([s], dtype=torch.float32, requires_grad=True)
        x_tensor = torch.tensor(x0[0], dtype=torch.float32, requires_grad=True)
        # print(x_tensor)
        #print("f_cv0(x_tensor, s_tensor, mu1)=",f_cv0(x_tensor, s_tensor, mu1))
        #print(s_tensor)
        return f_cv0(x_tensor, s_tensor, mu1) * f_cv0(x_tensor, s_tensor, mu2) / lbd / torch.exp(-lbd * s_tensor )

    # Integrand for single function
    def integrand_single1(s):
        s_tensor = torch.tensor([s], dtype=torch.float32, requires_grad=True)
        x_tensor = torch.tensor(x0[0], dtype=torch.float32, requires_grad=True)
        return f_cv0(x_tensor, s_tensor, mu1)
    def integrand_single2(s):
        s_tensor = torch.tensor([s], dtype=torch.float32, requires_grad=True)
        x_tensor = torch.tensor(x0[0], dtype=torch.float32, requires_grad=True)
        return f_cv0(x_tensor, s_tensor, mu2)

    # Numerical integration (discretized)
    s = torch.linspace(0, T.item(), steps=100)
    #print([integrand(s_i)[0] for s_i in s])
    integrand_values = torch.stack([integrand(s_i)[0] for s_i in s])
    # print(integrand_values)
    integral = torch.trapezoid(integrand_values, s)
    #print(integral)
    integrand_values_single1 = torch.stack([integrand_single1(s_i)[0] for s_i in s])
    #print(integrand_values_single1)
    integral_single1 = torch.trapezoid(integrand_values_single1, s)
    #print(integral_single1)
    integrand_values_single2 = torch.stack([integrand_single2(s_i)[0] for s_i in s])
    #print(integrand_values_single2)
    integral_single2 = torch.trapezoid(integrand_values_single2, s)
    #print(integral_single2)
    x_tensor = torch.tensor(x0[0], dtype=torch.float32, requires_grad=True)
    return (
        phi(x_tensor,mu1)*phi(x_tensor,mu2) / torch.exp(-lbd * T)
        + integral - (phi(x_tensor,mu1) + integral_single1)*(phi(x_tensor,mu2) + integral_single2)
    )

# Execute the cov function
#print(cov(x0=x0, T=T, mu1=mu, mu2=mu0))
#print(cov(x0=x0, T=T, mu1=mu0, mu2=mu0))
cstar = - cov(x0=x0, T=T, mu1=mu, mu2=mu0)/cov(x0=x0, T=T, mu1=mu0, mu2=mu0)
cstar=cstar.item()
print("cstar:", cstar)
covar = cov(x0=x0, T=T, mu1=mu, mu2=mu0)
print("covar:", covar.item())
corr = cov(x0=x0, T=T, mu1=mu, mu2=mu0)/torch.sqrt(cov(x0=x0, T=T, mu1=mu, mu2=mu)*cov(x0=x0, T=T, mu1=mu0, mu2=mu0))
print("corr^2:", corr.item() ** 2)
if np.isnan(cstar): cstar=-1 
cstar=-1
```

```python
# function definition
# deriv_map = np.array([0] * dim).reshape(-1, dim)
T=T.item()

deriv_map = np.concatenate((np.array([[0] * dim]), 1 * np.identity(dim), 2 * np.identity(dim)), axis=0).astype(int)

def f_example(y, coordinate=0):
    """
    idx 0 -> no deriv
    """
    # return ((gamma1-gamma2**2)/2/dim*((y[0:dim]-y[0:dim].sum(dim=0)/dim)**2).sum(dim=0) 
    #       + (mu+gamma2) * ((y[0:dim].sum(dim=0)/dim-y[0:dim])*y[dim:2*dim]).sum(dim=0) 
    #       - dim/2 * (y[dim:2*dim]**2).sum(dim=0))
    y1, y2 = y[:dim], y[1+dim:1+2*dim]; 
    return ((gamma1 - gamma2**2) / (2*dim) * ((y1.sum(dim=0)/dim-y1)**2).sum(dim=0) 
            + (mu + gamma2) * ((y1.sum(dim=0)/dim - y1)*y2).sum(dim=0)
            - dim/2 * (y2*y2).sum(dim=0))
    # return gamma2*2/dim*y[0:dim]**2.sum(dim=0) - dim/2/gamma1 * y[dim] * y[dim:2*dim].sum(dim=0) 

def f_example_cv(y, coordinate=0):
    """
    idx 0 -> no deriv
    """
    #return ((gamma1-gamma2**2)/2/dim*((y[0:dim]-y[0:dim].sum(dim=0)/dim)**2).sum(dim=0) 
    #        + (mu0+gamma2) * ((y[0:dim].sum(dim=0)/dim-y[0:dim])*y[dim:2*dim]).sum(dim=0) 
    #        - dim/2 * (y[dim:2*dim]**2).sum(dim=0))
    y1, y2 = y[:dim], y[1+dim:1+2*dim]; 
    return ((gamma1 - gamma2**2) / (2*dim) * ((y1.sum(0)/dim-y1)**2).sum(dim=0)  
            + (mu + gamma2) * ((y1.sum(0)/dim - y1)*y2).sum(dim=0)
            - dim/2 * (y2*y2).sum(dim=0))
#    return ((gamma1 - gamma2**2) / (2*n) * ((y1*y1).sum(0) - s1*s1/n) + 
#            (mu0 + gamma2) * (s1*s2/n - (y1*y2).sum(0)) - 
#            n/2 * (y2*y2).sum(0))
    # return y[0:dim].sum(dim=0) - mu0 * y[dim] * y[dim:2*dim].sum(dim=0) 
    # return - mu0 * y[0] * y[1:(dim+1)].sum(dim=0) 

# dgm_deriv_map = np.concatenate((deriv_map, 2 * np.identity(dim)), axis=0)
# dgm_deriv_map = np.concatenate((np.array([[0] * dim]), 1 * np.identity(dim), 2 * np.identity(dim)), axis=0).astype(int)
# dgm_deriv_map = np.concatenate((1 * np.identity(dim), 2 * np.identity(dim)), axis=0).astype(int)
# dgm_deriv_map = deriv_map
dgm_deriv_map = np.concatenate((1 * np.identity(dim), 2 * np.identity(dim)), axis=0).astype(int)

def dgm_f_example(y, coordinate=0):
    """
    idx 0      -> no deriv
    idx 1 to n -> second deriv
    """
#    return ((gamma1-gamma2**2)/2/dim*((y[0:dim]-y[0:dim].sum(dim=0)/dim)**2).sum(dim=0) 
#            + (mu+gamma2) * ((y[0:dim].sum(dim=0)/dim-y[0:dim])*y[dim:2*dim]).sum(dim=0) 
#            - dim/2 * (y[dim:2*dim]**2).sum(dim=0)
#            + nu * y[2*dim:3*dim].sum(dim=0) / 2)
    y1, y2 = y[:dim], y[dim:2*dim]; 
    return ((gamma1 - gamma2**2) / (2*dim) * ((y1.sum(dim=0)/dim-y1)**2).sum(dim=0) 
            + (mu + gamma2) * ((y1.sum(dim=0)/dim - y1)*y2).sum(dim=0)
            - dim/2 * (y2*y2).sum(dim=0)
            + nu * y[2*dim:3*dim].sum(dim=0) / 2)

# return - mu * y[0] * y[1:(dim+1)].sum(dim=0) + nu * y[(dim+1):(2*dim+1)].sum(dim=0) / 2 

def phi_example(x, coordinate=0):
    # return gamma3/2/dim*((x[0:dim].sum(dim=0)/dim-x[0:dim])**2).sum(dim=0)                       
#    x_first = x[0:dim].contiguous(); x_mean = x_first.mean(dim=0)
#    diff = x_mean - x_first  # Compute once, reuse
    return (gamma3 / (2 * dim) * ((x.sum(dim=0)/dim-x)**2).sum(dim=0) 
            + zzeta * gamma3 / (2 * dim) * ((x.sum(dim=0)/dim-x) ** 4).sum(dim=0))
    # return u2+(u1-u2)/(1+torch.exp((u1-u2)*mu*x.sum(dim=0)/nu))                       

def phi_example_2(x, coordinate=0):
    return (gamma3 / (2 * dim) * ((x.sum(axis=0)/dim-x)**2).sum(axis=0) 
            + zzeta * gamma3 / (2 * dim) * ((x.sum(axis=0)/dim-x) ** 4).sum(axis=0))

def phi_example_cv(x, coordinate=0):
    return gamma3/2/dim*((x[0:dim].sum(dim=0)/dim-x[0:dim])**2).sum(dim=0)                       
    # return u2+(u1-u2)/(1+torch.exp((u1-u2)*mu0*x.sum(dim=0)/nu))                       

# def phi_example(x, coordinate=0):
#    return -0.5 - 0.5 * torch.nn.Tanh()(-x[0] / 2)

# def exact_example(t, x, T):
#    return -0.5 - 0.5 * np.tanh(-x[0]/2 + 3*(T-t)/4)

delta=np.sqrt((mu+gamma2)**2+gamma1-gamma2**2)
delta0=np.sqrt((mu0+gamma2)**2+gamma1-gamma2**2)
M=gamma3+mu+gamma2-delta
M0=gamma3+mu0+gamma2-delta0

def alpha(t,T):
    return (delta-(mu+gamma2))/2/dim+1/dim/((2/M+1/delta)*np.exp(2*delta*(T-t))-1/delta)

def beta(t,T):
#    return nu*dim/2*(1-1/dim)*((delta-mu-gamma2)*(T-t)+np.log(1+M/2/delta*(1-np.exp(-2*delta*(T-t)))))
    return (1/(2*dim)*(dim-1)*nu*((gamma2 + mu + delta)*(t - T) 
           + np.log(-(gamma2 + gamma3 + mu -np.exp(2*delta*(-t + T))*(gamma2 + gamma3 + mu) 
           - delta - np.exp(2*delta*(-t + T))*delta)/2/delta)))
    
def exact_example(t, x, T):
    return (beta(t,T)+alpha(t,T)*((x-x.sum(axis=0)/dim)**2).sum(axis=0))                       

def alpha_2(t,T):
    return (delta-(mu+gamma2))/2/dim+1/(dim*((2/M+1/delta)*torch.exp(2*delta*(T-t))-1/delta))

def beta_2(t,T):
#    return nu*dim/2*(1-1/dim)*((delta-mu-gamma2)*(T-t)+torch.log(1+M/2/delta*(1-torch.exp(-2*delta*(T-t)))))
    return (1/(2*dim)*(dim-1)*nu*((gamma2 + mu + delta)*(t - T) 
           + torch.log(-(gamma2 + gamma3 + mu - torch.exp(2*delta*(-t + T))*(gamma2 + gamma3 + mu) 
           - delta - torch.exp(2*delta*(-t + T))*delta)/2/delta)))

def exact_example_2(t, x, T, coordinate=0):
    t = torch.tensor(t, dtype=torch.float32) if not isinstance(t, torch.Tensor) else t
    x = torch.tensor(x, dtype=torch.float32) if not isinstance(x, torch.Tensor) else x
    T = torch.tensor(T, dtype=torch.float32) if not isinstance(T, torch.Tensor) else T
    return (beta_2(t,T)+alpha_2(t,T)*((x-x.sum(dim=0)/dim)**2).sum(axis=0))                       

def boundary_fun(tx, coordinate):
    if coordinate == 0:
        return exact_example_2(tx[0], tx[1:], T)
    else:
        return exact_example_2(tx[0], tx[1:], T)
    # return torch.sin(tx[1]) * torch.cos(tx[2]) * torch.exp(-nu * (T - tx[0]))

def alpha_3(t,T):
    return (delta0-(mu0+gamma2))/2/dim+1/(dim*((2/M0+1/delta0)*torch.exp(2*delta0*(T-t))-1/delta0))

def beta_3(t,T):
#    return nu*dim/2*(1-1/dim)*((delta0-mu0-gamma2)*(T-t)+torch.log(1+M0/2/delta0*(1-torch.exp(-2*delta0*(T-t)))))
    return (1/(2*dim)*(dim-1)*nu*((gamma2 + mu0 + delta0)*(t - T) 
           + torch.log(-(gamma2 + gamma3 + mu0 - torch.exp(2*delta0*(-t + T))*(gamma2 + gamma3 + mu0) 
           - delta0 - torch.exp(2*delta0*(-t + T))*delta0)/2/delta0)))

def exact_example_cv(t, x, T, coordinate=0):
    #print(x)
    t = torch.tensor(t, dtype=torch.float32) if not isinstance(t, torch.Tensor) else t
    x = torch.tensor(x, dtype=torch.float32) if not isinstance(x, torch.Tensor) else x
    T = torch.tensor(T, dtype=torch.float32) if not isinstance(T, torch.Tensor) else T
    #print(t.size())
    #print("cv t=",t)
    #print(x.size())
    #print("cv x=",x)
    return (beta_3(t,T)+alpha_3(t,T)*((x-x.sum(dim=0)/dim)**2).sum(dim=0))                       

def alpha_4(t,T):
    return (delta0-(mu0+gamma2))/2/dim+1/(dim*((2/M0+1/delta0)*np.exp(2*delta0*(T-t))-1/delta0))

def beta_4(t,T):
#    return nu*dim/2*(1-1/dim)*((delta0-mu0-gamma2)*(T-t)+np.log(1+M0/2/delta0*(1-np.exp(-2*delta0*(T-t)))))
    return (1/(2*dim)*(dim-1)*nu*((gamma2 + mu0 + delta0)*(t - T) 
           + np.log(-(gamma2 + gamma3 + mu0 - np.exp(2*delta0*(-t + T))*(gamma2 + gamma3 + mu0) 
           - delta0 - np.exp(2*delta0*(-t + T))*delta0)/2/delta0)))

def exact_example_cv_2(t, x, T, coordinate=0):
    return (beta_4(t,T)+alpha_4(t,T)*((x-x.sum(axis=0)/dim)**2).sum(axis=0))                       
    # return u2+(u1-u2)/(1+np.exp((u1-u2)*(mu0*x.sum(axis=0)-mu0**2*(u1+u2)*(T-t)*dim/2)/nu))                       

grid = np.linspace(x_lo, x_hi, 100)
x_mid = (x_lo + x_hi) / 2
grid_d_dim = np.concatenate((x_mid * np.ones((fixed_dim, 100)), np.expand_dims(grid, axis=0), x_mid * np.ones((dim - fixed_dim - 1, 100))), axis=0)
grid_d_dim_with_t = np.concatenate((t_lo * np.ones((1, 100)), grid_d_dim), axis=0)

true = exact_example(t_lo, grid_d_dim, T)
#terminal = exact_example(T, grid_d_dim, T)
terminal = phi_example_2(grid_d_dim, coordinate=0)
true_nn = exact_example_cv_2(0, grid_d_dim, T)
```
