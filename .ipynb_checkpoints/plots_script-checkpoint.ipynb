```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Directory containing CSV files
data_dir = Path("logs_try_4")
#data_dir = Path("logs_L40_2")

# Output directory
output_dir = Path("plots_4")
#output_dir = Path("plots_3")
output_dir.mkdir(exist_ok=True)

# List of files
files = [
    "allen_cahn_dim_1_time_behavior.csv",
    "allen_cahn_dim_10_time_behavior.csv",
    "allen_cahn_dim_100_time_behavior.csv",
    "allen_cahn_dim_1000_time_behavior.csv",
    "allen_cahn_2_dim_1_time_behavior.csv",
    "allen_cahn_2_dim_10_time_behavior.csv",
    "allen_cahn_2_dim_100_time_behavior.csv",
    "allen_cahn_2_dim_1000_time_behavior.csv",
    "allen_cahn_3_dim_1_time_behavior.csv",
    "allen_cahn_3_dim_10_time_behavior.csv",
    "allen_cahn_3_dim_100_time_behavior.csv",
    "allen_cahn_3_dim_1000_time_behavior.csv",
    "exp_example_dim_1_time_behavior.csv",
    "exp_example_dim_10_time_behavior.csv",
    "exp_example_dim_100_time_behavior.csv",
    "exp_example_dim_1000_time_behavior.csv",
]

for file in files:

    path = data_dir / file
    if not path.exists():
        print(f"Skipping missing file: {file}")
        continue

    problem_name = path.stem

    # Read CSV
    df = pd.read_csv(path)

    # Replace 'nan' strings if present
    df = df.replace("nan", float("nan"))

    # Convert columns to float
    cols = ["T", "branch", "NPP23", "bsde", "exact"]
    for c in cols:
        df[c] = pd.to_numeric(df[c], errors="coerce")

    # Plot
    fig, ax = plt.subplots(figsize=(8,5))
    # fig, ax = plt.subplots()

    ax.plot(df['T'], df['branch'], linestyle='-', color='red', label='Binary')
    ax.plot(df['T'], df['NPP23'], color='blue', label='NPP23')
    if not problem_name.startswith("allen_cahn_3"):
        ax.plot(df['T'], df['exact'], color='black', linestyle='--', label='Exact')
    # ax.plot(df['T'], df['exact'], color='black', linestyle='--', label='Exact')

    ax.plot(df['T'], df['bsde'], color='purple', label='BSDE')
    
    ax.plot(df['T'], df['NPP23'], linestyle='None', marker='.', markersize=8,
            color='blue', markerfacecolor='none')

    ax.plot(df['T'], df['bsde'], linestyle='None', marker='.', markersize=8,
            color='purple', markerfacecolor='none')

    ax.plot(df['T'], df['branch'], linestyle='None', marker='o', markersize=8,
            color='red', markerfacecolor='none')

    ax.set_xlabel("T")
    ax.margins(x=0)

    if problem_name.startswith("allen_cahn_dim"):
        legend_loc = "lower left"
    elif (
        problem_name.startswith("allen_cahn_2")
        or problem_name.startswith("exp_example")
        or problem_name == "allen_cahn_3_dim_1_time_behavior"
    ):
        legend_loc = "upper left"
    else:
        legend_loc = "upper right"

    ax.legend(loc=legend_loc, fontsize=16)

    #    ax.legend(loc="upper right")

    ax.grid(True)

    plt.tight_layout()

    output_path = output_dir / f"{problem_name}.pdf"
    fig.savefig(output_path, bbox_inches="tight")
    plt.close(fig)

    print(f"Saved {output_path}")
```
