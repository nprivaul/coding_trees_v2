```python
def blow_up_analysis(T_arr, dim, problem_name, deriv_map, cstar, f_example, f_example_cv, phi_example, phi_example_cv, exact_example, exact_example_cv, nu):
    df = []
    exact = exact_example(t=0, x=np.zeros((dim, 1)), T=0, coordinate=0)[0]
    df.append([0, exact, exact, exact, exact, 0, 0, 0])
    for T in T_arr:
        print(f"Now dealing with T = {T}...")
        # initialize branching model
        torch.manual_seed(0)
        model = Net(
            problem_name=problem_name,
            f_fun=f_example,
            deriv_map=deriv_map,
            phi_fun=phi_example,
            device=device,
            x_lo=0.0,
            x_hi=0.0,
            nu=nu,
            T=T,
            verbose=True,
            branch_nb_path_per_state=100000,
            branch_nb_states=1,
            branch_nb_states_per_batch=1,
            save_as_tmp=True,
        )
        print("Model complete")
        #model_cv = Net_cv(
        #    problem_name=problem_name,
        #    cstar=cstar,
        #    f_fun=f_example,
        #    f_fun_cv=f_example_cv,
        #    exact_fun_cv=exact_example_cv,
        #    deriv_map=deriv_map,
        #    phi_fun=phi_example,
        #    phi_fun_cv=phi_example_cv,
        #    device=device,
        #    x_lo=1.0,
        #    x_hi=1.0,
        #    nu=nu,
        #    T=T,
        #    verbose=True,
        #    branch_nb_path_per_state=1000,
        #    save_as_tmp=True,
        #)
        
        # tx = torch.zeros(1, model.dim_in + 1, device=device)
        tx = None
        
        # branching mc calculation
        start = time.time()
        mc, repeats = 0, 1
        for _ in range(repeats):
            x, y = model.gen_sample(patch=0, tx=tx)
            mc += y[:, 0].cpu()
            print("1000 samples")
        mc = mc[0].item() / repeats
        mc_time = time.time() - start

        start = time.time()
        mc_cv, repeats = 0, 10
        #for _ in range(repeats):
        #    x, y = model_cv.gen_sample_cv(patch=0, tx=tx)
        #    mc_cv += y[:, 0].cpu()
        #mc_cv = mc_cv[0].item() / repeats
        mc_cv_time = time.time() - start
        
        # exact calculation & bsde calculation
        exact = exact_example(t=0, x=np.zeros((model.dim_in, 1)), T=T, coordinate=0)[0]
        start = time.time()
#        bsde = deep_bsde(total_time=T, verbose=False, nu=nu, num_iterations= 50, dim=dim, y_init_range=[0,1])[-1][2]
        bsde = deep_bsde(total_time=T, verbose=False, nu=nu, num_iterations= 1000, dim=dim, y_init_range=[0,0.1])[-1][2]
        bsde_time = time.time() - start
        # exact = 0.61061
        df.append([T, mc, mc_cv, bsde, exact, mc_time, mc_cv_time, bsde_time])
        ddf = pd.DataFrame(df, columns=['T', 'branch', 'branch_cv', 'deep_bsde', 'exact', 'branch time (seconds)', 'branch_cv time (seconds)', 'deep_bsde time (seconds)'])
        print(ddf)       
        
    # create df & save
    if not os.path.isdir('logs/'):
        os.makedirs('logs/')
    df = pd.DataFrame(df, columns=['T', 'branch', 'branch_cv', 'deep_bsde', 'exact', 'branch time (seconds)', 'branch_cv time (seconds)', 'deep_bsde time (seconds)'])
    df.to_csv("logs/" + problem_name + ".csv")
    
    return df
```

```python
fixed_dim, dim = 0, 10
T = torch.tensor(0.1 , dtype=torch.float32)  # Time horizon
mu0 = 0.0
mu = 0.0 # mu0+0.1*dim

lbd = torch.tensor(1.0, dtype=torch.float32)  # Lambda parameter

t_lo, x_lo, x_hi = 0., 0., 0.
x_mid = (x_lo + x_hi) / 2
# x0 = x_mid * np.eye(1, dim) 
x0 = 1.0 * np.eye(1, dim) 

u1=0.
u2=1.

nu=2
lambd = (u1 + u2) / 2

def nth_derivatives(order, y, x):
    """
    Compute nth derivative of y with respect to x.
    """
    for _ in range(order):
        try:
            grads = torch.autograd.grad(y.sum(), x, create_graph=True)[0]
        except RuntimeError:
            # Handle zero gradient case gracefully
            return torch.zeros_like(y)
        y = grads
    return y

def u_cv(x, T):
    return u2+(u1-u2)/(1+torch.exp((u1-u2)*(mu0*x.sum(dim=0)-mu0**2*(u1+u2)*T*dim/2)/nu))                       

def u_cv_der(y, T):
    """
    First derivative of u_cv with respect to y.
    """
    return nth_derivatives(1, u_cv(y, T), y)

def f_cv0(y, T, mu):
    u_prime = u_cv_der(y, T)
    return (
        - mu * u_cv(y, T) * u_prime
    )
    
def phi(x,mu):
    """
    Example phi function.
    """
    return u2+(u1-u2)/(1+torch.exp((u1-u2)*(mu0*x.sum(dim=0))/nu))                       

def cov(x0, T, mu1, mu2): 
    """
    Compute covariance.
    """
    # Define the integrand for numerical integration
    def integrand(s):
        s_tensor = torch.tensor([s], dtype=torch.float32, requires_grad=True)
        x_tensor = torch.tensor(x0[0], dtype=torch.float32, requires_grad=True)
        # print(x_tensor)
        #print("f_cv0(x_tensor, s_tensor, mu1)=",f_cv0(x_tensor, s_tensor, mu1))
        #print(s_tensor)
        return f_cv0(x_tensor, s_tensor, mu1) * f_cv0(x_tensor, s_tensor, mu2) / lbd / torch.exp(-lbd * s_tensor )

    # Integrand for single function
    def integrand_single1(s):
        s_tensor = torch.tensor([s], dtype=torch.float32, requires_grad=True)
        x_tensor = torch.tensor(x0[0], dtype=torch.float32, requires_grad=True)
        return f_cv0(x_tensor, s_tensor, mu1)
    def integrand_single2(s):
        s_tensor = torch.tensor([s], dtype=torch.float32, requires_grad=True)
        x_tensor = torch.tensor(x0[0], dtype=torch.float32, requires_grad=True)
        return f_cv0(x_tensor, s_tensor, mu2)

    # Numerical integration (discretized)
    s = torch.linspace(0, T.item(), steps=100)
    #print([integrand(s_i)[0] for s_i in s])
    integrand_values = torch.stack([integrand(s_i)[0] for s_i in s])
    # print(integrand_values)
    integral = torch.trapezoid(integrand_values, s)
    #print(integral)
    integrand_values_single1 = torch.stack([integrand_single1(s_i)[0] for s_i in s])
    #print(integrand_values_single1)
    integral_single1 = torch.trapezoid(integrand_values_single1, s)
    #print(integral_single1)
    integrand_values_single2 = torch.stack([integrand_single2(s_i)[0] for s_i in s])
    #print(integrand_values_single2)
    integral_single2 = torch.trapezoid(integrand_values_single2, s)
    #print(integral_single2)
    x_tensor = torch.tensor(x0[0], dtype=torch.float32, requires_grad=True)
    return (
        phi(x_tensor,mu1)*phi(x_tensor,mu2) / torch.exp(-lbd * T)
        + integral - (phi(x_tensor,mu1) + integral_single1)*(phi(x_tensor,mu2) + integral_single2)
    )

# Execute the cov function
#print(cov(x0=x0, T=T, mu1=mu, mu2=mu0))
#print(cov(x0=x0, T=T, mu1=mu0, mu2=mu0))
cstar = - cov(x0=x0, T=T, mu1=mu, mu2=mu0)/cov(x0=x0, T=T, mu1=mu0, mu2=mu0)
cstar=cstar.item()
print("cstar:", cstar)
covar = cov(x0=x0, T=T, mu1=mu, mu2=mu0)
print("covar:", covar.item())
corr = cov(x0=x0, T=T, mu1=mu, mu2=mu0)/torch.sqrt(cov(x0=x0, T=T, mu1=mu, mu2=mu)*cov(x0=x0, T=T, mu1=mu0, mu2=mu0))
print("corr^2:", corr.item() ** 2)
if np.isnan(cstar): cstar=-1 
cstar=-1
```

```python
# function definition
# deriv_map = np.array([0] * dim).reshape(-1, dim)
T=T.item()

gamma1=1.0; gamma2=0.1; gamma3=0.1;

# deriv_map = np.concatenate((np.array([[0] * dim]), 1 * np.identity(dim), 2 * np.identity(dim)), axis=0).astype(int)
# deriv_map = np.concatenate((1 * np.identity(dim), 2 * np.identity(dim)), axis=0).astype(int)
deriv_map = np.concatenate((np.array([[0] * dim]), 1 * np.identity(dim), 2 * np.identity(dim)), axis=0).astype(int)

def f_example(y, coordinate=0):
    """
    idx 0 -> no deriv
    """
    return (gamma2/2/dim*(y[0:dim]**2).sum(dim=0) 
            - mu * (y[0:dim]*y[dim:2*dim]).sum(dim=0) 
            - dim/2/gamma1 * (y[dim:2*dim]**2).sum(dim=0))
    # return gamma2*2/dim*y[0:dim]**2.sum(dim=0) - dim/2/gamma1 * y[dim] * y[dim:2*dim].sum(dim=0) 

def f_example_cv(y, coordinate=0):
    """
    idx 0 -> no deriv
    """
    return (gamma2/2/dim*(y[0:dim]**2).sum(dim=0) 
            - mu0 * (y[0:dim]*y[dim+1:2*dim+1]).sum(dim=0) 
            - dim/2/gamma1 * (y[dim+1:2*dim+1]**2).sum(dim=0))
    # return y[0:dim].sum(dim=0) - mu0 * y[dim] * y[dim:2*dim].sum(dim=0) 
    # return - mu0 * y[0] * y[1:(dim+1)].sum(dim=0) 

dgm_deriv_map = np.concatenate((deriv_map, 2 * np.identity(dim)), axis=0)

def dgm_f_example(y, coordinate=0):
    """
    idx 0      -> no deriv
    idx 1 to n -> second deriv
    """
    return gamma2*2/dim*(y[0:dim]**2).sum(dim=0) - dim/2/gamma1 * (y[dim+1:2*dim+1]**2).sum(dim=0) + nu * y[dim+1:2*dim+1].sum(dim=0) / 2
    # return - mu * y[0] * y[1:(dim+1)].sum(dim=0) + nu * y[(dim+1):(2*dim+1)].sum(dim=0) / 2 

def phi_example(x, coordinate=0):
    return gamma3/2/dim*(x[0:dim]**2).sum(dim=0)                       
#    return u2+(u1-u2)/(1+torch.exp((u1-u2)*mu*x.sum(dim=0)/nu))                       

def phi_example_cv(x, coordinate=0):
    return gamma3/2/dim*(x[0:dim]**2).sum(dim=0)                       
    # return u2+(u1-u2)/(1+torch.exp((u1-u2)*mu0*x.sum(dim=0)/nu))                       

# def phi_example(x, coordinate=0):
#    return -0.5 - 0.5 * torch.nn.Tanh()(-x[0] / 2)

# def exact_example(t, x, T):
#    return -0.5 - 0.5 * np.tanh(-x[0]/2 + 3*(T-t)/4)

lambdas=0.5*(np.sqrt(gamma2/gamma1)-gamma3/gamma1)

def exact_example(t, x, T, coordinate=0):
    sigmas2 = (1-np.exp(-2*np.sqrt(gamma2/gamma1)*(T-t)))/2/np.sqrt(gamma2/gamma1)
    return (np.sqrt(gamma1*gamma2)*(T-t)
            + gamma1*np.log(1 - 2 * sigmas2 * lambdas) 
            + (np.sqrt(gamma1*gamma2)+2*gamma1*lambdas*np.exp(-2*np.sqrt(gamma2/gamma1)*(T-t))
            / (2*lambdas*sigmas2-1))*(x**2).sum(axis=0)/2/dim)                       
    # return u2+(u1-u2)/(1+np.exp((u1-u2)*(mu*x.sum(axis=0)-mu**2*(u1+u2)*(T-t)*dim/2)/nu))                       

def exact_example_cv(t, x, T, coordinate=0):
    #print(x)
    t = torch.tensor(t, dtype=torch.float32) if not isinstance(t, torch.Tensor) else t
    x = torch.tensor(x, dtype=torch.float32) if not isinstance(x, torch.Tensor) else x
    T = torch.tensor(T, dtype=torch.float32) if not isinstance(T, torch.Tensor) else T
    sigmas2 = (1-torch.exp(-2*np.sqrt(gamma2/gamma1)*(T-t)))/2/np.sqrt(gamma2/gamma1)
    #print(t.size())
    #print("cv t=",t)
    #print(x.size())
    #print("cv x=",x)
    return (np.sqrt(gamma1*gamma2)*(T-t)
            + gamma1*torch.log(1 - 2 * sigmas2* lambdas) 
            + (np.sqrt(gamma1*gamma2)+2*gamma1*lambdas*torch.exp(-2*np.sqrt(gamma2/gamma1)*(T-t))
            / (2*lambdas*sigmas2-1))*(x**2).sum(dim=0)/2/dim)                          
    # return u2+(u1-u2)/(1+torch.exp((u1-u2)*(mu0*x.sum(dim=0)-mu0**2*(u1+u2)*(T-t)*dim/2)/nu))                       

def exact_example_cv_2(t, x, T, coordinate=0):
    sigmas2 = (1-np.exp(-2*np.sqrt(gamma2/gamma1)*(T-t)))/2/np.sqrt(gamma2/gamma1)
    return (np.sqrt(gamma1*gamma2)*(T-t)
            + gamma1*np.log(1 - 2 * sigmas2* lambdas) 
            + (np.sqrt(gamma1*gamma2)+2*gamma1*lambdas*np.exp(-2*np.sqrt(gamma2/gamma1)*(T-t))
            / (2*lambdas*sigmas2-1))*(x**2).sum(axis=0)/2/dim)                          
    # return u2+(u1-u2)/(1+np.exp((u1-u2)*(mu0*x.sum(axis=0)-mu0**2*(u1+u2)*(T-t)*dim/2)/nu))                       

grid = np.linspace(x_lo, x_hi, 100)
x_mid = (x_lo + x_hi) / 2
grid_d_dim = np.concatenate((x_mid * np.ones((fixed_dim, 100)), np.expand_dims(grid, axis=0), x_mid * np.ones((dim - fixed_dim - 1, 100))), axis=0)
grid_d_dim_with_t = np.concatenate((t_lo * np.ones((1, 100)), grid_d_dim), axis=0)

true = exact_example(t_lo, grid_d_dim, T)
terminal = exact_example(T, grid_d_dim, T)
true_nn = exact_example_cv_2(0, grid_d_dim, T)
```
