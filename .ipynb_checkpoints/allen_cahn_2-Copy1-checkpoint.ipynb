## Allen Cahn Equation

```python
problem_name = f"allen_cahn_2_dim_{dim}_time_behavior"

deriv_map = np.array([0] * dim).reshape(-1, dim)

mu=1

def f_example(y):
    """
    idx 0 -> no deriv
    """
    return mu * ( y[0] - y[0] ** 3 )

def phi_example(x, coordinate=0):
    scale = torch.sqrt(torch.tensor(x.shape[0], dtype=x.dtype, device=x.device))
    z = x.sum(dim=0) / scale
    z = -0.1 * z
    return 0.1 - 0.0 * torch.tanh(z)

def exact_example(t, x, T, coordinate=0):
    x = torch.tensor(x, dtype=torch.float32) if not isinstance(x, torch.Tensor) else x
    return np.ones_like(x) * 1/np.sqrt(1-(1-0.1**(-2))*np.exp(-2*(T-t)))

T=0.5
delta_t = 0.05
T_arr = np.arange(delta_t, T + delta_t, delta_t)
branch_nb_path_per_state=10000
branch_nb_states_per_batch=10
branch_nb_states=1000
outlier_percentile=1
repeats=100
num_iterations=4000
y_init_range=[0,1]
xlo=0.0
xhi=0.0
nu=1.0
```

### Dimension d = 1

```python
df = blow_up_analysis(
    T_arr=T_arr,
    fixed_dim=0,
    dim=1,
    problem_name=problem_name,
    deriv_map=deriv_map,
    f_example=f_example,
    phi_example=phi_example,
    exact_example=exact_example,
    branch_nb_path_per_state=branch_nb_path_per_state, 
    branch_nb_states_per_batch=branch_nb_states_per_batch,
    branch_nb_states=branch_nb_states,
    outlier_percentile=outlier_percentile,
    repeats=repeats,
    nu=nu,
    num_iterations=num_iterations,
    y_init_range=y_init_range,
    xlo=xlo,
    xhi=xhi
)
```
