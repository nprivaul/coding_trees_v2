```python
import re
from pathlib import Path
from collections import defaultdict

#root = Path("coding_trees/logs_L40_2")
root = Path("coding_trees/logs_try_4")

# Problems to include
target_problems = {"allen_cahn", "allen_cahn_2", "allen_cahn_3", "exp_example"}

# Filename pattern
pattern = re.compile(
    r'^(?P<problem>allen_cahn(?:_[23])?|exp_example)_dim_(?P<dim>\d+)_time_behaviortable\.tex$'
)

results = defaultdict(dict)

# ---- Scan files ----
for file in root.glob("*_time_behaviortable.tex"):
    match = pattern.match(file.name)
    if not match:
        continue

    problem = match.group("problem")
    dim = int(match.group("dim"))

    if problem not in target_problems:
        continue

    with open(file) as f:
        lines = f.readlines()

    last_row = None
    for line in lines:
        if line.strip().startswith("0.5"):
            last_row = line.strip().rstrip("\\")
    if last_row:
        last_row = last_row.replace("nan", "NaN")
        results[problem][dim] = last_row        

# ---- Write LaTeX files ----
output_dir = Path("latex_tables_4")
output_dir.mkdir(exist_ok=True)

for problem, dim_data in results.items():
    output_file = output_dir / f"{problem}_summary.tex"

    # Determine header and column format
    if problem == "allen_cahn_3":
        header = "Dim & Binary & NPP23 & BSDE & Binary (s) & NPP23 (s) & BSDE (s)"
        ncols = header.count("&") + 1
        remove_exact = True
    else:
        header = "Dim & Binary & NPP23 & BSDE & Exact & Binary (s) & NPP23 (s) & BSDE (s)"
        ncols = header.count("&") + 1
        remove_exact = False

    col_format = "c " * ncols

    with open(output_file, "w") as f:
        f.write(f"\\begin{{tabular}}{{{col_format}}}\n")
        f.write("\\hline\n")
        f.write(header + " \\\\\n")
        f.write("\\hline\n")

        for dim in sorted(dim_data):
            original_row = dim_data[dim]
            row_parts = original_row.split("&")[1:]  # skip original T=0.5

            if remove_exact:
                # remove the Exact column (4th item: index 3)
                if len(row_parts) >= 4:
                    del row_parts[3]

            new_row = f"{dim} & " + " & ".join(part.strip() for part in row_parts)
            f.write(new_row + " \\\\\n")

        f.write("\\hline\n")
        f.write("\\end{tabular}\n")

    print(f"Saved: {output_file}")
```
